```python
from random import randrange, seed
#import random
import numpy as np
import matplotlib.pyplot as plt

seed(123456) 
```

```python
file1 = open("input_vectors.txt","w")

nsamples = 1000
fs = 500
frequency = 1
x = np.arange(nsamples)
signal1 = np.sin(2 * np.pi * frequency * x / fs)

v = []

for i in signal1:
    p = np.int(i * np.power(2,6) + np.power(2,6) )
    file1.write(str(p))
    file1.write("\n")
    v.append(p)
    #print(p)

file1.close()

#print(signal1)
#plt.plot(signal1)
#plt.show()
```

```python
c = [1,2,5,2,1]

o = []

for i in range(5,len(v)):
    a = 0
    for j in range(4):
        a += v[i-j]*c[j]
    o.append(a)
    
v = np.array(v)
o = np.array(o)
```

```python
print(o.max()/v.max())
```

```python
plt.plot(np.arange(0,len(v)),v,'o',markersize=0.2)
plt.plot(np.arange(0,len(o)),o,'o',markersize=0.2)
```
